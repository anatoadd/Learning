```python
import random

def check_hit_and_blow(secret, guess):
    """ユーザーの推測値と正解を比較して、ヒットとブローの数を返す"""

    # ヒットとブロー変数の初期化
    hit = 0
    blow = 0

# 全角で入力された場合に半角へ変換する関数
def　to_half_width(s):
    return s.translate(str.maketrans('０１２３４５６７８９', '0123456789'))

    #　ヒットカウント(ヒット=数字と位置が合っている)
    for i in range(len(secret)):
        if secret[i] == guess[i]:
            hit += 1

    # 重複数のカウント
    hit_and_blow = 0
    for num in secret:
        if num in guess:
            hit_and_blow += 1

    # ブロー = 重複数からhit数を引く
    blow = hit_and_blow - hit

    return hit,blow

# ゲーム開始の説明
print('数当てゲームスタート')
print('私が1~9までの数値を使ってランダムな数を作ります。')
print('あなたは1桁から9桁の桁数を指定してください。')

# 桁数入力
while True:
    n = input('何桁の数字でゲームしますか？(1~9):')
    n = to_half_width(n)
    if n.isdigit() and 1 <= int(n) <= 9:
        n = int(n)
        break
    print('1~9の数字を入力してください。')

# 正解の数
numbers = list(range(1,10))
secret_numbers = random.sample(numbers, n)

# 試行回数を初期化
trial_count = 0

# ユーザーから推測した数字を受け取って正解するまでループを回す
while True:
    guess_number = input(f'{n}桁の数字を入力してください。')
    guess_number = to_half_width(guess_number)

    ## 入力の長さを確認
    if len(guess_number) != n or not guess_number.isdigit():
        print(f'入力文字数が不正です。{n}桁の数字を入力してください。')
        continue

    # 入力を整数のリストに変換
    guess_list = [int(char) for char in guess_number]

    # 試行回数をカウントアップ
    trial_count += 1
    print(f'{trial_count}回目の回答です。')

    # ユーザーの推測値を正解と比較し、hit数とblow数を返す
    hit, blow = check_hit_and_blow(secret_numbers, guess_list)

    # 結果表示
    if hit == n:
        print(f'正解!　ゲームクリアです!　正解 = {secret_numbers}')
        print(f'{trial_count}回で正解しました。')
        break
    else:
        print(f'ヒット = {hit}, ブロー = {blow}')
```
